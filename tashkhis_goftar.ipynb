 تبدیل سیگنال‌های صوتی به حوزه فرکانس

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# خواندن فایل صوتی
#  add to main code in book آپلود فایل
print("لطفاً فایل صوتی خود را آپلود کنید (مثال: random_sound.wav):")
uploaded = files.upload()

# بررسی نام فایل آپلود شده
file_name = list(uploaded.keys())[0]
print(f"فایل آپلود شده: {file_name}")

#return name file  خواندن فایل صوتی
sampling_freq, signal = wavfile.read(file_name)

# نرمال‌سازی مقادیر
signal = signal / np.power(2, 15)

#  (tedade kole nemone ha dar signale soti)استخراج طول سیگنال صوتی
len_signal = len(signal)
# استخراج نیمه طول
len_half = np.ceil((len_signal + 1) / 2.0).astype(int)

# اعمال تبدیل فوریه
freq_signal = np.fft.fft(signal)

# نرمال‌سازی سیگنال فرکانسی
freq_signal = abs(freq_signal[0:len_half]) / len_signal
# محاسبه مربع سیگنال فرکانسی
freq_signal **= 2

# استخراج طول سیگنال تبدیل فوریه
len_fts = len(freq_signal)
# (ask)تنظیم سیگنال برای حالات زوج و فرد
if len_signal % 2:
    freq_signal[1:len_fts] *= 2
else:
    freq_signal[1:len_fts-1] *= 2

# استخراج قدرت سیگنال به دسی‌بل
signal_power = 10 * np.log10(freq_signal)

# (bande ferekanse /tole signal)ساخت محور X (فرکانس به کیلوهرتز)
x_axis = np.arange(0, len_half, 1) * (sampling_freq / len_signal) / 1000.0

# رسم نمودار
plt.figure()
plt.plot(x_axis, signal_power, color='hotpink')
plt.xlabel('فرکانس (کیلوهرتز)')
plt.ylabel('قدرت سیگنال (دسی‌بل)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from google.colab import drive
drive.mount('/content/drive')

# فایل خروجی که صدا در آن ذخیره خواهد شد
output_file = '/content/drive/MyDrive/generated_audio.wav'


# مشخص کردن پارامترهای سیگنال صوتی
duration = 4  # مدت زمان (ثانیه)
sampling_freq = 44100  # فرکانس نمونه‌برداری (هرتز)
tone_freq = 784  # فرکانس تن (هرتز)
min_val = -4 * np.pi  # مقدار حداقل برای محور زمان
max_val = 4 * np.pi  # مقدار حداکثر برای محور زمان

# تولید سیگنال صوتی
t = np.linspace(min_val, max_val, duration * sampling_freq)
signal = np.sin(2 * np.pi * tone_freq * t)

# افزودن نویز به سیگنال
noise = 0.5 * np.random.rand(duration * sampling_freq)
signal += noise

# نرمال‌سازی و مقیاس‌بندی سیگنال
scaling_factor = np.power(2, 15) - 1
signal_normalized = signal / np.max(np.abs(signal))
signal_scaled = np.int16(signal_normalized * scaling_factor)

# ذخیره سیگنال صوتی در فایل خروجی
write(output_file, sampling_freq, signal_scaled)

# استخراج 200 مقدار اول از سیگنال برای رسم
signal = signal[:200]

# ساخت محور زمان در میلی‌ثانیه
time_axis = 1000 * np.arange(0, len(signal), 1) / float(sampling_freq)

# رسم نمودار سیگنال صوتی
plt.plot(time_axis, signal, color='hotpink')
plt.xlabel('Time (milliseconds)')
plt.ylabel('Amplitude')
plt.title('Generated audio signal')
plt.show()
```

سنتز کردن نت‌ها برای تولید موسیقی

```python
import json  # برای کار با داده‌های JSON
import numpy as np  # برای محاسبات عددی و کار با آرایه‌ها
from scipy.io.wavfile import write  # برای نوشتن فایل‌های WAV

#تعریف تابع تولید صدا
def tone_synthesizer(freq, duration, amplitude=1.0, sampling_freq=44100):
    num_samples = int(duration * sampling_freq)  # محاسبه تعداد نمونه‌ها بر اساس مدت زمان و فرکانس نمونه‌برداری
    t = np.linspace(0, duration, num_samples, endpoint=False)  # ایجاد محور زمان از 0 تا مدت زمان مشخص شده

    # ساخت سیگنال صوتی با استفاده از تابع سینوس
    signal = amplitude * np.sin(2 * np.pi * freq * t)

    return signal.astype(np.int16)  # تبدیل نوع داده به int16 برای ذخیره‌سازی در فایل WAV

if __name__ == '__main__':  # بررسی اینکه آیا این فایل به عنوان برنامه اصلی اجرا می‌شود
    # نام فایل‌های خروجی
    file_tone_single = '/content/drive/MyDrive/generated1_audio.wav'  # فایل برای صدای تک
    file_tone_sequence = '/content/drive/MyDrive/generated2_audio.wav'  # فایل برای توالی صداها

    # مسیر فایل JSON که شامل نقشه فرکانس به صدا است
    mapping_file = '/content/drive/MyDrive/tone_mapping.json'

    # بارگذاری نقشه فرکانس به صدا از فایل JSON
    with open(mapping_file, 'r') as f:  # باز کردن فایل JSON برای خواندن
        tone_map = json.loads(f.read())  # خواندن محتوای فایل و تبدیل آن به دیکشنری

    # تنظیم پارامترهای ورودی برای تولید صدای 'F'
    tone_name = 'F'  # نام صدایی که می‌خواهیم تولید کنیم
    duration = 3     # مدت زمان صدای تولید شده (ثانیه)
    amplitude = 12000  # شدت صدای تولید شده
    sampling_freq = 44100  # فرکانس نمونه‌برداری (هرتز)

    # استخراج فرکانس مربوط به صدای 'F' از نقشه فرکانس‌ها
    tone_freq = tone_map[tone_name]

    # تولید صدای بر اساس پارامترهای بالا
    synthesized_tone = tone_synthesizer(tone_freq, duration, amplitude, sampling_freq)

    # نوشتن سیگنال صوتی در فایل خروجی
    write(file_tone_single, sampling_freq, synthesized_tone)  # ذخیره صدای تولید شده در فایل WAV

    # تعریف توالی صدا همراه با مدت زمان‌های مربوطه در ثانیه
    tone_sequence = [('G', 0.4), ('D', 0.5), ('F', 0.3), ('C', 0.6), ('A', 0.4)]  # توالی صداها و مدت زمان آنها

    # ساخت سیگنال صوتی بر اساس توالی بالا
    signal = np.array([], dtype=np.int16)  # ایجاد آرایه خالی برای نگهداری سیگنال نهایی
    for item in tone_sequence:  # تکرار بر روی هر عنصر در توالی صدا
        tone_name = item[0]  # دریافت نام صدای فعلی

        # استخراج فرکانس مربوط به صدای فعلی از نقشه
        freq = tone_map[tone_name]

        # استخراج مدت زمان صدای فعلی
        duration = item[1]

        # تولید صدای فعلی
        synthesized_tone = tone_synthesizer(freq, duration, amplitude, sampling_freq)

        # اضافه کردن سیگنال خروجی به سیگنال نهایی
        signal = np.concatenate((signal, synthesized_tone))  # ترکیب سیگنال‌های تولید شده

    # ذخیره سیگنال نهایی در فایل خروجی
    write(file_tone_sequence, sampling_freq, signal)  # ذخیره توالی صداها در فایل WAV
```

```python
import os
import sys
import argparse
import warnings
import numpy as np
from scipy.io import wavfile
from hmmlearn import hmm
from python_speech_features import mfcc
from google.colab import drive


drive.mount('/content/drive')

def build_arg_parser():
    parser = argparse.ArgumentParser(description='Trains the HMM-based speech recognition system')
    parser.add_argument("--input-folder", dest="input_folder", required=True,
                        help="Input folder containing the audio files for training")
    return parser

class ModelHMM(object):
    def __init__(self, num_components=4, num_iter=1000):
        self.n_components = num_components
        self.n_iter = num_iter
        self.cov_type = 'diag'
        self.model_name = 'GaussianHMM'
        self.models = []
        self.model = hmm.GaussianHMM(n_components=self.n_components,
                                     covariance_type=self.cov_type, n_iter=self.n_iter)

    def train(self, training_data):
        np.seterr(all='ignore')
        cur_model = self.model.fit(training_data)
        self.models.append(cur_model)

    def compute_score(self, input_data):
        return self.model.score(input_data)


def build_models(input_folder):
    speech_models = []

    for dirname in os.listdir(input_folder):
        subfolder = os.path.join(input_folder, dirname)

        if not os.path.isdir(subfolder):
            continue


        label = os.path.basename(subfolder)

        X = np.array([])

        training_files = [x for x in os.listdir(subfolder) if x.endswith('.wav') and not x.startswith('15')]


        for filename in training_files:
            filepath = os.path.join(subfolder, filename)
            sampling_freq, signal = wavfile.read(filepath)


            with warnings.catch_warnings():
                warnings.simplefilter('ignore')
                features_mfcc = mfcc(signal, sampling_freq)

            if len(X) == 0:
                X = features_mfcc
            else:
                X = np.append(X, features_mfcc, axis=0)


        model = ModelHMM()
        model.train(X)
        speech_models.append((model, label))

    return speech_models


def run_tests(test_files, speech_models):
    for test_file in test_files:
        sampling_freq, signal = wavfile.read(test_file)

        with warnings.catch_warnings():
            warnings.simplefilter('ignore')
            features_mfcc = mfcc(signal, sampling_freq)

        max_score = -float('inf')
        predicted_label = None


        for model, label in speech_models:
            score = model.compute_score(features_mfcc)
            if score > max_score:
                max_score = score
                predicted_label = label


        original_label = os.path.basename(os.path.dirname(test_file))
        print(f'\nOriginal: {original_label}')
        print(f'Predicted: {predicted_label}')


if __name__ == '__main__':

    sys.argv = ['script_name', '--input-folder', '/content/drive/MyDrive/data']
    args = build_arg_parser().parse_args()
    input_folder = args.input_folder


    speech_models = build_models(input_folder)


    test_files = []
    for root, dirs, files in os.walk(input_folder):
        for filename in (x for x in files if '15' in x and x.endswith('.wav')):
            filepath = os.path.join(root, filename)
            test_files.append(filepath)

    run_tests(test_files, speech_models)
```

```python
import speech_recognition as sr
from gtts import gTTS
import os
from google.colab import files
from pydub import AudioSegment
import requests  # Importing requests to connect to the API for Persian data

# Upload MP3 file
uploaded = files.upload()

# Get the name of the uploaded file
audio_file = next(iter(uploaded))

# Convert MP3 file to WAV
audio = AudioSegment.from_mp3(audio_file)
audio.export("file.wav", format="wav")

# Function to speak using Google TTS (gTTS)
def speak(text):
    tts = gTTS(text=text, lang='en')
    tts.save("speech.mp3")
    os.system("mpg321 speech.mp3")  # This command plays the generated speech.mp3 file

# Function to listen from an audio file and convert to text
def listen_from_audio(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        print("Listening...")
        audio = recognizer.record(source)  # Read the entire file
        print("Recognizing...")
        try:
            # Use Google Speech Recognition API for Persian language
            query = recognizer.recognize_google(audio, language='fa-IR')  # Specify Persian language ('fa-IR')
            print(f"You said: {query}")
            return query.lower()  # Convert to lowercase for easier comparison
        except sr.UnknownValueError:
            print("Sorry, I didn't understand that.")
            return None
        except sr.RequestError:
            print("Sorry, there was an issue with the speech service.")
            return None

# Connect to a Persian dataset API (replace URL with your Persian data API)
def get_persian_data(query):
    # Replace this with the URL of your actual Persian data API
    api_url = f'https://example.com/persian_data?query={query}'  # Replace with your actual API URL

    try:
        # Send GET request to the API
        response = requests.get(api_url)

        if response.status_code == 200:
            # Assuming the API returns a JSON with a field "data"
            data = response.json()
            return data.get("data", "No data found for your query.")
        else:
            return "Sorry, I couldn't retrieve data from the API."
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return "Failed to connect to the API."

# Process the audio file
command = listen_from_audio("file.wav")

# If the command is understood, fetch related Persian data
if command:
    print(f"Processing query: {command}")
    persian_data = get_persian_data(command)
    speak(f"The related data is: {persian_data}")
else:
    speak("Sorry, I couldn't understand anything.")
```

```python
import jiwer
from jiwer import wer

def evaluate_model_accuracy(transcriptions, ground_truths):
    """
    Evaluates the accuracy of speech recognition models using WER.

    Parameters:
    transcriptions (list): List of transcriptions generated by the model.
    ground_truths (list): List of reference (ground truth) transcriptions.

    Returns:
    float: The average Word Error Rate (WER).
    """
    if len(transcriptions) != len(ground_truths):
        raise ValueError("Number of transcriptions must match the number of ground truths")

    # Compute WER for each transcription
    total_wer = 0
    for pred, truth in zip(transcriptions, ground_truths):
        total_wer += wer(truth, pred)

    # Compute the average WER
    avg_wer = total_wer / len(transcriptions)
    return avg_wer

# Example usage
transcriptions = [
    "this is a test",
    "hello world",
    "speech recognition is cool"
]

ground_truths = [
    "this is a test",
    "hello world",
    "speech recognition is awesome"
]

# Evaluate accuracy
average_wer = evaluate_model_accuracy(transcriptions, ground_truths)
print(f"Average Word Error Rate (WER): {average_wer:.2f}")
```
